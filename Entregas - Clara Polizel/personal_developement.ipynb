```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
```

```python
load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
for model in genai.list_models():
    print(model.name)
```

```python
def gerar_resposta_gemini(prompt):
    model = genai.GenerativeModel("models/gemini-1.5-flash")
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
def criar_rotina_pessoal(objetivo, tempo_disponivel, preferencias):
    prompt = f"""
    Voc√™ √© um coach especializado em desenvolvimento pessoal. Crie uma rotina di√°ria personalizada 
    para uma pessoa cujo objetivo √©: {objetivo}.
    
    Considere:
    - Tempo dispon√≠vel por dia: {tempo_disponivel}
    - Prefer√™ncias da pessoa: {preferencias}
    
    A rotina deve incluir:
    - Hor√°rios estimados (ex: manh√£, tarde, noite)
    - Atividades sugeridas (ex: leitura, exerc√≠cio, estudo, medita√ß√£o)
    - Dicas para manter const√¢ncia e motiva√ß√£o
    """
    return gerar_resposta_gemini(prompt)
```

```python
def analisar_habitos(habitos_atuais):
    prompt = f"""
    Voc√™ √© um especialista em mudan√ßa de h√°bitos. Analise os h√°bitos abaixo e:
    - Identifique os que s√£o prejudiciais
    - Sugira h√°bitos substitutos saud√°veis
    - D√™ dicas pr√°ticas de como implementar e manter esses novos h√°bitos
    
    H√°bitos informados:
    {habitos_atuais}
    """
    return gerar_resposta_gemini(prompt)
```

```python
objetivo = "melhorar a concentra√ß√£o e criar uma rotina de estudos"
tempo_disponivel = "6 horas por dia (manh√£ e tarde)"
preferencias = "gosta de estudar com m√∫sica baixa, fazer pausas curtas, e prefere estudar sozinho"

habitos_atuais = """
- Procrastina frequentemente
- Usa o celular por longos per√≠odos antes de dormir
- N√£o faz pausas regulares ao estudar
- N√£o tem hor√°rio fixo para acordar
"""
```

```python
rotina = criar_rotina_pessoal(objetivo, tempo_disponivel, preferencias)
analise = analisar_habitos(habitos_atuais)

print("üìÖ Rotina Sugerida:\n")
print(rotina)
print("\nüîç An√°lise de H√°bitos e Recomenda√ß√µes:\n")
print(analise)
```
